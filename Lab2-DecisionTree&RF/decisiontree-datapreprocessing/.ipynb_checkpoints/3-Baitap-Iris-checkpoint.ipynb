# Bài tập Decision Tree

## Sử dụng Decision Tree để phân nhóm cho tập Iris dataset
- dataset: 'Iris.csv'


```python
import pandas as pd
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.keys()
```

```python
df.head()
```

```python
# column Id and Species is no needed (Id does not bring any meaning in training, while Species is the label), so I drop them.
# The X variable is our attribute set and y variable contains corresponding labels.
X = df.drop(["Id", "Species"], axis=1)
y = df["Species"]
```

```python
X.head()
```

```python
y.head()
```

```python
# change label into number 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

```python
# SPLIT INTO TRAINING SET & TEST SET
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
print(X_train.shape, y_train.size)
print(X_test.shape, y_test.shape)
```

```python
# BUILD DECISION TREE WITH 'GINI' AS CRITERION
from sklearn.tree import DecisionTreeClassifier  
dt = DecisionTreeClassifier(criterion='gini')  
dt.fit(X_train, y_train)  
```

```python
y_pred_dt = dt.predict(X_test)  
y_pred_dt
```

```python
dt_score = dt.score(X_test, y_test)
print(f"Decision Tree classifier accuracy score is {dt_score}")
```

```python
# BUILD DECISION TREE WITH 'ENTROPY' AS CRITERION
from sklearn.tree import DecisionTreeClassifier  
dt2 = DecisionTreeClassifier(criterion='entropy')  
dt2.fit(X_train, y_train)  
```

```python
y_pred_dt = dt2.predict(X_test)  
y_pred_dt
```

```python
dt_score = dt.score(X_test, y_test)
print(f"Decision Tree classifier accuracy score is {dt_score}")
```

```python
# VISUALIZE DECISION TREE
from sklearn.tree import export_graphviz
dot_data = export_graphviz(dt, out_file=None)
print(dot_data)
```

```python
import graphviz 
graph = graphviz.Source(dot_data)
graph
```
