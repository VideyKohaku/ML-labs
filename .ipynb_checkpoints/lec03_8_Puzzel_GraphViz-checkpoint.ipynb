## Defining Node class

Complete TODOs 1 - 3

```python
from copy import deepcopy
from graphviz import Digraph

class Node:

  def __init__(self, state, action = None, parent = None):
    self.state = state # 2D list (3x3)
    self.id = str(self) # identifier of node
    self.action = action
    self.parent = parent

  def __str__(self):
    '''
    # TODO 1
    Return a string representing the state of the node.
    Note to remember the 0 digit by '_'
    e.g., '182\n_43\n765'
    '''
    return '\n'.join([''.join([str(digit) if digit != 0 else '_' for digit in row]) for row in self.state])

  def get_successors(self):
    '''
    # TODO 2
    Return the list of successors of the state (self).
    '''
    action_lst = ['L', 'R', 'U', 'D']
    successors_lst = []
    for action in action_lst:
      successor_state = self.get_successor(action, deepcopy(self.state))
      if successor_state is not None:
          if self.parent is None or self.parent.state != successor_state:
            successors_lst.append(Node(successor_state, action, parent=self))
    return successors_lst

    
  def get_successor(self, action, state):
    pi, pj = self.get_blank_pos(state)
    pi, pj = self.get_dest_pos(action, pi, pj)
    if 0 <= pi and pi < 3 and 0 <= pj and pj < 3:
      if action == 'L':
        state[pi][pj - 1] = state[pi][pj]
      if action == 'R':
        state[pi][pj + 1] = state[pi][pj]
      if action == 'U':
        state[pi-1][pj] = state[pi][pj]
      if action == 'D':
        state[pi+1][pj] = state[pi][pj]
      state[pi][pj] = 0
      return state
    return None

  def get_dest_pos(self, action, pi, pj):
    if action == 'L':
      pj += 1
    if action == 'R':
      pj -= 1
    if action == 'U':
      pi += 1
    if action == 'D':
      pi -= 1
    return pi, pj

  def get_blank_pos(self, state):
    '''
    # TODO 3
    Return the location (i, j) of the 0 digit (blank cell).
    '''
    for pi in range(len(state)):
        for pj in range(len(state[pi])):
            if state[pi][pj] == 0:
                return pi, pj

  def get_id(self):
    return self.id

  def get_node_str(self):
    return str(self)

  def get_action(self):
    return self.action

  def draw(self, dot):
    dot.node(self.get_id(), self.get_node_str())
    if self.parent is not None:
      dot.edge(self.parent.get_id(), self.get_id(), self.get_action())
```

```python
def expand(dot, n):
  successors = n.get_successors()
  for s in successors:
    s.draw(dot)

n = Node([[1, 8, 2],
          [0, 4, 3],
          [7, 6, 5]])
dot = Digraph()
n.draw(dot)
expand(dot, n)

dot
```

## Implement graph-search

Complete TODO 4

```python
def graph_search(node, max_num_nodes = 20):
  dot = Digraph()
  frontier = list()
  explored = set()
  frontier.append(node)
  count = 0

  while len(frontier) > 0 and count < max_num_nodes:
    '''
    # TODO 4
    Complete the body of the loop statement regarding the graph-search algorithm.
    The output is an example of the tree expanded the the process.
    '''
    # pop the frontier
    current_node = frontier.pop(0)
    # add node to explored set
    explored.add(current_node.get_id())


    # expand the current_node
    successors = current_node.get_successors()
    for successor in successors:
        if successor.get_id() not in explored:
            frontier.append(successor)
    count += len(successors)
    if count <= max_num_nodes:
        expand(dot, current_node)

  return dot

n = Node([[1, 8, 2],
          [0, 4, 3],
          [7, 6, 5]])
dot = graph_search(n, 33)
dot
```
